```python
import matplotlib.pyplot as plt

def bre(x0,y0,x1,y1):
    x=[]
    y=[]
    dx=abs(x1-x0)
    dy=abs(y1-y0)
    err=dx-dy

    if x0<x1:
        sx=1
    else:
        sx=-1

    if y0<y1:
        sy=1
    else:
        sy=-1

    while x0 != x1 or y0!=y1:
        x.append(x0)
        y.append(y0)

        e2=2*err

        if e2>-dy:
            err-=dy
            x0+=sx
        if e2<dx:
            err+=dx
            y0+=sy

    x.append(x1)
    y.append(y1)

    return x,y

def main():
    plt.figure()

    x,y=bre(20,40,40,60)
    plt.plot(x,y)
    x,y=bre(20,80,40,60)
    plt.plot(x,y,'.')
    x,y=bre(20,40,40,20)
    plt.plot(x,y,'.')
    x,y=bre(60,80,80,80)
    plt.plot(x,y)
    x,y=bre(60,20,60,80)
    plt.plot(x,y)
    x,y=bre(60,20,80,20)
    plt.plot(x,y)
    x,y=bre(80,20,80,80)
    plt.plot(x,y)
    x,y=bre(100,40,120,60)
    plt.plot(x,y,'.')
    x,y=bre(100,40,120,20)
    plt.plot(x,y)
    x,y=bre(100,80,120,60)
    plt.plot(x,y)
    x,y=bre(140,60,170,60)
    plt.plot(x,y)
    x,y=bre(160,20,160,70)
    plt.plot(x,y)
    x,y=bre(140,90,140,60)
    plt.plot(x,y)


    plt.show()



if __name__== "__main__":
    main()
```
