```python

import matplotlib.pyplot as plt

def draw_circle(xc,yc,x,y):
    plt.plot(xc+x,yc+y,'.')
    plt.plot(xc-x,yc+y,'.')
    plt.plot(xc+x,yc-y,'.')
    plt.plot(xc-x,yc-y,'.')
    plt.plot(xc+y,yc+x,'.')
    plt.plot(xc-y,yc+x,'.')
    plt.plot(xc+y,yc-x,'.')
    plt.plot(xc-y,yc-x,'.')


def mid(xc,yc,r):
    x=0
    y=r
    p=1-r
    draw_circle(xc,yc,x,y)

    while x<y:
        x+=1
        if p<0:
            p+=2*x+1
        else:
            y-=1
            p+=2*(x-y)+1

        draw_circle(xc,yc,x,y)


def bre(x0,y0,x1,y1):
    x=[]
    y=[]

    dx=abs(x1-x0)
    dy=abs(y1-y0)
    err=dx-dy

    if x0<x1:
        sx=1
    else:
        sx=-1

    if y0<y1:
        sy=1
    else:
        sy=-1

    while x0 != x1 or y0!=y1:
        x.append(x0)
        y.append(y0)

        e2=2*err

        if e2>-dy:
            err-=dy
            x0+=sx
        if e2<dx:
            err+=dx
            y0+=sy

    x.append(x1)
    y.append(y1)

    return x,y

def main():
    plt.figure()

    x,y=bre(10,30,120,30)
    plt.plot(x,y)
    x,y=bre(20,20,110,20)
    plt.plot(x,y,'.')
    x,y=bre(30,10,100,10)
    plt.plot(x,y)
    x,y=bre(10,30,20,20)
    plt.plot(x,y)
    x,y=bre(20,20,30,10)
    plt.plot(x,y)
    x,y=bre(100,10,110,20)
    plt.plot(x,y)
    x,y=bre(110,20,120,30)
    plt.plot(x,y,'.')

    # upper top

    x,y=bre(30,30,30,110)
    plt.plot(x,y)
    x,y=bre(30,110,40,110)
    plt.plot(x,y)
    x,y=bre(40,30,40,110)
    plt.plot(x,y)
    x,y=bre(80,30,80,110)
    plt.plot(x,y)
    x,y=bre(80,110,90,110)
    plt.plot(x,y)
    x,y=bre(90,30,90,110)
    plt.plot(x,y)

    # middle
    x,y=bre(50,30,50,50)
    plt.plot(x,y)
    x,y=bre(60,30,60,50)
    plt.plot(x,y)
    x,y=bre(70,30,70,50)
    plt.plot(x,y)
    x,y=bre(50,90,50,110)
    plt.plot(x,y)
    x,y=bre(60,90,60,110)
    plt.plot(x,y)
    x,y=bre(70,90,70,110)
    plt.plot(x,y)

    # upper middle

    x,y=bre(50,110,60,120)
    plt.plot(x,y)
    x,y=bre(60,110,70,120)
    plt.plot(x,y)
    x,y=bre(70,110,80,120)
    plt.plot(x,y)
    x,y=bre(60,120,80,120)
    plt.plot(x,y)

#    circle
    mid(60,70,20)

    plt.show()



if __name__== "__main__":
    main()
```
